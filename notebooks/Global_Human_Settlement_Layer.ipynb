# Global Human Settlement Layer

[Global Human Settlement Layer](https://human-settlement.emergency.copernicus.eu/) (GHSL) is a global [geospatial raster](https://geopandas.org/en/stable/gallery/geopandas_rasterio_sample.html) TIF data set that describes "the human presence on Earth."

```python
from pathlib import Path
import matplotlib.pyplot
import rasterio.plot
```

```python
DATA_DIR = Path("../data")
```

# Load GHSL maps

In Python, we can use TIF files using [Rasterio](https://rasterio.readthedocs.io/).

```python
# Iterate over GHSL files
for path in DATA_DIR.glob("GHS_*.tif"):
    # Open raster file
    with rasterio.open(path) as ghsl:
        # Plot
        fig, ax = matplotlib.pyplot.subplots()
        rasterio.plot.show(ghsl, ax=ax, cmap="magma")
        ax.set_title(path.name)
        ax.set_xlabel("Longitude")
        ax.set_ylabel("Latitude")
```
